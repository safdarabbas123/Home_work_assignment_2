# Show stacked bar chart by categories
## A simple stacked bar chart of all components of primary energy supply for one scenario.

```python
import pyam

df = pyam.IamDataFrame("AR6_Scenarios_Database_World_v1.1.csv")
df
```

```python
args = dict(model="WITCH-GLOBIOM 4.4", scenario="CD-LINKS_NPi2020_1000")
data = df.filter(**args, variable="Primary Energy|*", region="World")

ax = data.plot.bar(stacked=True, title="Primary energy mix")
legend_labels = [var.split("|")[1] for var in data.columns]
ax.legend(labels=legend_labels, loc=1)
plt.tight_layout()
plt.show()
```

# Stacked line charts

```python
model, scenario = "IMAGE 3.0.1", "CD-LINKS_NPi2020_400"

data = df.filter(
    model=model, scenario=scenario, variable="Primary Energy|*", region="World"
)

ax = data.plot.stack(title=scenario)
legend_labels = [var.split("|")[1] for var in data.columns]
ax.legend(labels=legend_labels, loc=0)
plt.tight_layout()
plt.show()
```
